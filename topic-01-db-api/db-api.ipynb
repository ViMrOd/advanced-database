```python
connection = sqlite3.connect("pets.db")
cursor = connection.cursor()
```

```python
cursor.execute("""
    create table if not exists pets (
        id integer primary key autoincrement,
        name text not null,
        type text not null,
        age integer,
        owner text
    )
""")

connection.commit()
```

```python
name = "dorothy"
cursor.execute("""
    insert 
        into pets(name, type, age, owner) 
        values (?,?,?,?)
    """,(name,"dog",9,"greg"))
connection.commit()
```
